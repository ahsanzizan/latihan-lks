#### Implementation of Breadth-First Search algorithm in Python

```python
import networkx as nx
import matplotlib.pyplot as plt
from collections import deque
```

```python
# Create a random graph
num_nodes = 10  # Number of nodes
probability = 0.3  # Probability of edge creation
G: nx.Graph = nx.erdos_renyi_graph(num_nodes, probability)

def draw_graph(G: nx.Graph, visited=set()):
    positions = nx.spring_layout(G, seed=3)
    nx.draw(G, pos=positions)
    nx.draw_networkx_nodes(G, pos=positions, nodelist=visited, node_color='red')
    nx.draw_networkx_labels(G, pos=positions, labels={node: str(node) for node in G.nodes})
    plt.show()
    
draw_graph(G)
```

```python
def process_node(node: str):
    print(f'Processing {node}')

def bfs(G: nx.Graph, start: str):
    visited = set()
    queue = deque([start])
    visited.add(start)
    
    while queue:
        current = queue.popleft()
        process_node(current)
        
        for neighbor in G.neighbors(current):
            if neighbor not in visited:
                draw_graph(G, visited)
                visited.add(neighbor)
                queue.append(neighbor)
```

```python
import random

print("Number of nodes:", G.number_of_nodes())
print("Number of edges:", G.number_of_edges())

start_node = random.choice(list(G.nodes))

print("\nBFS traversal starting from node", start_node)
bfs(G, start_node)
```
