#### Implementation of Depth-First Search algorithm in Python

```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G = nx.Graph()

G.add_edge('A', 'B')
G.add_edge('B', 'C')
G.add_edge('A', 'C')
G.add_edge('C', 'D')
G.add_edge('A', 'D')
G.add_edge('D', 'E')
G.add_edge('C', 'E')
G.add_edge('B', 'E')
G.add_edge('A', 'F')
G.add_edge('B', 'F')

def draw_graph(G: nx.Graph, visited=set()):
    positions = nx.spring_layout(G, seed=1)
    nx.draw(G, pos=positions)
    nx.draw_networkx_nodes(G, pos=positions, nodelist=visited, node_color='red')
    nx.draw_networkx_labels(G, pos=positions, labels={node: str(node) for node in G.nodes})
    plt.show()
    
draw_graph(G)
```

```python
def process_node(node: str):
    print(f'Processing {node}')

def dfs(G: nx.Graph, start: str, visited=None):
    if visited is None:
        visited = set()
        
    visited.add(start)
    process_node(start) # Process the node, in this case print it
    
    for neighbor in G.neighbors(start):
        if neighbor not in visited:
            draw_graph(G, visited)
            dfs(G, neighbor, visited)
    
```

```python
# Define a starting node
start_node = 'A'
dfs(G, start_node)
```
