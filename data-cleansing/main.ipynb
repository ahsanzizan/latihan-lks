# Get The Missing Data Points Count

```python
# Get the missing data points
missing_values = nfl_data.isnull()

# Get the count of missing data points
missing_values_count = missing_values.sum()

missing_values_count[:10]
```

### Percentage of Missing Data Points

```python
# Total missing values
total_cells = np.product(nfl_data.shape)
total_missing = missing_values_count.sum()

# Percentage of data that is missing
percent_missing = (total_missing / total_cells) * 100

percent_missing
```

# Drop Missing Values

It looks like it removed all of our data. This is because every row in our dataset had at least one missing value.

```python
nfl_data.dropna()
```

```python
# Remove all columns with at least one missing value
data_with_na_dropped = nfl_data.dropna(axis=1)
data_with_na_dropped.head()
```

```python
# How much columns did we lose?
print(f"Columns in original dataset: {nfl_data.shape[1]}")
print(f"Columns with NA dropped: {data_with_na_dropped.shape[1]}")
```

# Filling in The Missing Values

```python
# Get a small subset of the NFL dataset
subset_nfl_data = nfl_data.loc[:, 'EPA' : 'Season'].head()
subset_nfl_data
```

#### Filling the NAs with 0

```python
# Replace all NAs with 0
subset_nfl_data.fillna(0)
```

#### Filling the NAs with the value that comes directly after it in the same column

```python
# replace all NA's the value that comes directly after it in the same column, 
# then replace all the remaining na's with 0
subset_nfl_data.bfill(axis=0).fillna(0)
```
