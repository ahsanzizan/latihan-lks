```python
import networkx as nx
```

```python
# Creating a graph
G = nx.Graph()
```

```python
def draw():
    nx.draw(G)
```

## Nodes

A Graph is a collection of nodes (vertices) along with identified pairs of nodes (edges). Nodes can be hashable objects e.g., a text string, an image, an XML object, another Graph, etc.

```python
G.add_node(1) # Adding a node
```

```python
draw()
```

```python
G.add_nodes_from([2, 3, 4])
```

```python
draw()
```

```python
G.add_nodes_from([(4, {"color": "red"}), (5, {"color": "green"})])
```

```python
draw()
```

Nodes from one graph can be incorporated into another:

```python
H = nx.path_graph(10)
G.add_nodes_from(H)
draw()
```

we could use the graph H as a node in G

```python
G.add_node(H)
```

```python
draw()
```

## Edges

`G` can also be grown by adding one edge at a time

```python
G.add_edges_from([(1, 2), (3, 4), (5, 1), (5, 3), (10, 1), (3, 1), (1, 4), (5, 2), (2, 9), (9, 10)])
draw()
```

or by adding any `ebunch` of edges. An `ebunch` is any iterable container of edge-tuples. An edge-tuple can be a 2-tuple of nodes or a 3-tuple with 2 nodes followed by an edge attribute dictionary, e.g., (2, 3, {'weight': 3.1415}).


```python
G.number_of_nodes()
```

```python
G.number_of_edges()
```

```python
G.clear() # Clearing the graph
```

## Directed Graph

```python
DG = nx.DiGraph()
```

```python
def draw():
    nx.draw(DG)
```

```python
DG.add_nodes_from(range(100, 110))
```

```python
H = nx.path_graph(10)
DG.add_nodes_from(H)
```

```python
draw()
```

```python
DG.add_edges_from([(1, 2), (3, 1), (1, 5), (5, 2), (10, 9), (9, 3), (10, 6), (5, 6)])
draw()
```

Each graph, node, and edge can hold key/value attribute pairs in an associated attribute dictionary (the keys must be hashable). By default these are empty, but can be added or changed using add_edge, add_node or direct manipulation of the attribute dictionaries named graph, node and edge respectively.
